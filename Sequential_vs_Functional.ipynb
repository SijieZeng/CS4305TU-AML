```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```
# Define Sequential model with 3 layers using the Class Constructor
model = keras.Sequential(
    [
        layers.Input(shape=(3,), name='layer0'),
        layers.Dense(2, activation="relu", name="layer1"),
        layers.Dense(3, activation="relu", name="layer2"),
        layers.Dense(4, name="layer3"),      
    ], name= 'Sequential Model'
)

# Model summary
model.summary()
```

```
# Define Sequential model with 3 layers using the .add method
model = keras.Sequential(name="Sequential Model")
model.add(layers.Input(shape=(3,), name='layer0'))
model.add(layers.Dense(2, activation="relu", name="layer1"))
model.add(layers.Dense(3, activation="relu", name="layer2"))
model.add(layers.Dense(4, name="layer3"))

# Model summary
model.summary()
```

```
# Define the same model using the Functional API
i = layers.Input(shape=(3,), name='layer0')
x = layers.Dense(2, activation="relu", name="layer1")(i)
x = layers.Dense(3, activation="relu", name="layer2")(x)
x = layers.Dense(4, name="layer3")(x)
model = keras.Model(inputs=i, outputs=x, name="Functional Model")

# Model summary
model.summary()
```

```
# Define branched model using the Functional API
i = layers.Input(shape=(3,), name='layer0')
x = layers.Dense(2, activation="relu", name="layer1")(i)
x = layers.Dense(3, activation="relu", name="layer2") (x)
b1 = layers.Dense(3, activation="relu", name="branch1") (x)
b2 = layers.Dense(3, activation="relu", name="branch2") (x)
b = layers.add([b1,b2], name="branch_merge")
x = layers.Dense(4, name="output")(b)
model = keras.Model(inputs=i, outputs=x, name="Branched Functional Model")

# Model summary
model.summary()
```

```
keras.utils.plot_model(model, "branched_model.png")
```
